```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager #크롬드라이버
from selenium.webdriver.common.by import By

def set_chrome_driver():
    chrome_options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
    return driver

url = 'https://www.kfckorea.com/store/findStore' # url 설정
driver = set_chrome_driver()
driver.get(url)
```

```python
import time # 코드 시간지연을 위한 라이브러리
from tqdm import tqdm_notebook # 현재 크롤링이 어느정도까지 지연됐는지 알수있음.
```

```python
driver.find_element(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[2]/div/div[2]/div/div/ul/li[2]/a').click()
do_list_raw = driver.find_elements(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[2]/div/div[2]/div/div/div/div/div[1]/select[1]')
do_list_raw = do_list_raw[0].text.split('\n')
do_list_raw
```

```python
si_list_raw = driver.find_elements(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[2]/div/div[2]/div/div/div/div/div[1]/select[2]')
si_list_raw = si_list_raw[0].text.split('\n')
si_list_raw
```

```python
from selenium.webdriver.support.select import Select
get_column = [] # 값을 저장할 리스트
try:
    for i in tqdm_notebook(do_list_raw):
        driver.find_element(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[2]/div/div[2]/div/div/div/div/div[1]/select[1]').click()
        driver.find_element(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[2]/div/div[2]/div/div/div/div/div[1]/select[1]').send_keys(i)
        driver.find_element(By.XPATH, '//*[@id="app"]/div[2]/div/div/div[2]/div/div[2]/div/div/div/div/div[1]/select[2]').click()
        si_list_raw = driver.find_elements(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[2]/div/div[2]/div/div/div/div/div[1]/select[2]')
        si_list_raw = si_list_raw[0].text.split('\n')
        time.sleep(0.4)
        for j in tqdm_notebook(range(len(si_list_raw))):
            select = Select(driver.find_element(By.XPATH, '//*[@id="app"]/div[2]/div/div/div[2]/div/div[2]/div/div/div/div/div[1]/select[2]'))
            select.select_by_index(j)
            time.sleep(0.4)
            for z in range(int(driver.find_element(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[2]/div/div[2]/div/div/div/div/p/span/em').text)):
                driver.find_element(By.XPATH, '//*[@id="app"]/div[2]/div/div/div[2]/div/div[2]/div/div/div/div/div[2]/div[1]/div/ul['+str(z+1)+']/li[1]/a').click()
                time.sleep(0.4)
                info = []            
                info.append(str(driver.find_element(By.XPATH, '//*[@id="app"]/div[2]/div/div/div[3]/div[2]/div/div/div[2]/h2').text))
                info.append(str(driver.find_elements(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[3]/div[2]/div/div/div[3]/p')[0].text))
                info.append(driver.find_element(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[3]/div[2]/div/div/div[3]/div/ul/li[1]/span').text)
                info.append(driver.find_element(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[3]/div[2]/div/div/div[3]/div/ul/li[2]/span').text)
                info.append(driver.find_element(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[3]/div[2]/div/div/div[3]/div/ul/li[3]/span').text)
                info.append(driver.find_element(By.XPATH,'//*[@id="app"]/div[2]/div/div/div[3]/div[2]/div/div/div[3]/div/ul/li[4]/span').text)
                driver.find_element(By.XPATH, '//*[@id="app"]/div[2]/div/div/div[3]/div[2]/div/a').click()
                time.sleep(0.4)
                get_column += info
except:
    pass
        
```
