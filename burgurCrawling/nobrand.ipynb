```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager #크롬드라이버
from selenium.webdriver.common.by import By

def set_chrome_driver():
    chrome_options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
    return driver

url = 'https://www.shinsegaefood.com/nobrandburger/store/store.sf' # url 설정
driver = set_chrome_driver()
driver.get(url)
```

```python
import time # 코드 시간지연을 위한 라이브러리
from tqdm import tqdm_notebook # 현재 크롤링이 어느정도까지 지연됐는지 알수있음.
```

```python
store_list = (driver.find_elements(By.XPATH, '//*[@id="typeCode02"]')[0].text).split('\n')
store_list
```

```python
from selenium.webdriver.support.select import Select
get_column = [] # 값을 저장할 리스트

for i in tqdm_notebook(range(len(store_list))):
    select = Select(driver.find_element(By.XPATH, '//*[@id="typeCode02"]'))
    select.select_by_index(i)
    time.sleep(0.2)
    get_column.append((driver.find_element(By.XPATH, '//*[@id="brand_middle"]/div/section/div[4]/div/div[2]/div[1]/h2').text))
    get_column.append((driver.find_element(By.XPATH, '//*[@id="store_story"]').text).split('\n'))
    time.sleep(0.1)
```

```python
get_column = [get_column[i:i+2] for i in range(0, len(get_column), 2)]
# 리스트를 5개 단위로 나눠주며 가공
```
