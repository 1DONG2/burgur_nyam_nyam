```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager #크롬드라이버
from selenium.webdriver.common.by import By

def set_chrome_driver():
    chrome_options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
    return driver

url = 'https://www.subway.co.kr/storeSearch?page=1&rgn1Nm=&rgn2Nm=#storeList' # url 설정
driver = set_chrome_driver()
driver.get(url)
```

```python
import time # 코드 시간지연을 위한 라이브러리
from tqdm import tqdm_notebook # 현재 크롤링이 어느정도까지 지연됐는지 알수있음.
```

```python
get_column = [] # 값을 저장할 리스트
for i in tqdm_notebook(range(1, 56)): # 지점목록이 56번까지 있음.
    # 해당 값을 가져온다.
    get_column += (driver.find_elements(By.XPATH, '//*[@id="boardWrap"]/ul')[0].text).split('\n')
    url = 'http://frankburger.co.kr/board/index.php?board=map_01&sca=all&type=list&select&search=&page='+str(i)
    driver.get(url) # 해당 사이트로 접속
    time.sleep(1) # 새로고침을 위해 코드 1초 지연
```

```python
get_column = [get_column[i:i+5] for i in range(0, len(get_column), 5)]
# 리스트를 5개 단위로 나눠주며 가공
```

```python
get_column
```

```python
#데이터 가공
for i in range(len(get_column)):
    get_column[i][0] = get_column[i][0].split('[')[0] # 점포 이름 가공
    get_column[i][1] = get_column[i][1][2:] # 주소 지우기
    get_column[i][2] = get_column[i][2][3:].replace(')','-') # 전화 )을 - 바꾸기 및 '전화 ' 없애기
    get_column[i][3] = get_column[i][3][2:] # 오픈 지우기
    get_column[i].remove('더보기') # 더보기 지우기
```

```python
for i in range(len(get_column)):
    if float(get_column[i][-1].replace(' ','').split('[')[0][0:4]) < 1000.0:
        get_column[i][-1] = '20'+get_column[i][-1].replace(' ','').split('[')[0]
```

```python
# 열심히 구한 데이터를 csv 파일로 저장하자
import csv
get_column
with open('C:\\Users\\우리집\\Downloads\\frank.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(get_column)
```
