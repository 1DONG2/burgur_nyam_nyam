```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager #크롬드라이버
from selenium.webdriver.common.by import By

def set_chrome_driver():
    chrome_options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
    return driver

url = 'https://www.burgerking.co.kr/#/store' # url 설정
driver = set_chrome_driver()
driver.get(url)
```

```python
import time # 코드 시간지연을 위한 라이브러리
from tqdm import tqdm_notebook # 현재 크롤링이 어느정도까지 지연됐는지 알수있음.
```

```python
driver.find_element(By.XPATH,'//*[@id="app"]/div/div[3]/div[3]/div/div[1]/div[1]/div[1]/ul/li[3]/button').click()
do_list_raw = driver.find_elements(By.XPATH,'//*[@id="app"]/div/div[3]/div[3]/div/div[1]/div[1]/div[2]/div[4]/div/select[1]')
do_list_raw = do_list_raw[0].text.split('\n')
do_list_raw
```

```python
from selenium.webdriver.support.select import Select
get_column = [] # 값을 저장할 리스트
try:
    for i in tqdm_notebook(range(1, len(do_list_raw))):
        select = Select(driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div[3]/div/div[1]/div[1]/div[2]/div[4]/div/select[1]'))
        select.select_by_index(i)
        time.sleep(0.4)
        driver.find_element(By.XPATH,'//*[@id="app"]/div/div[3]/div[3]/div/div[1]/div[2]/div/p/em/span').click()
        for z in tqdm_notebook(range(int(driver.find_element(By.XPATH,'//*[@id="app"]/div/div[3]/div[3]/div/div[1]/div[2]/div/p/em/span').text))):
            driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div[3]/div/div[1]/div[2]/ul/li['+str(z+1)+']/button').click()
            time.sleep(0.4)
            info = []            
            info.append(str(driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div[4]/div/div[1]/div/h1/span').text))
            info.append(driver.find_elements(By.XPATH,'//*[@id="app"]/div/div[3]/div[4]/div/div[2]/div/div/dl[1]/dd/span[1]')[0].text)
            info.append(driver.find_element(By.XPATH,'//*[@id="app"]/div/div[3]/div[4]/div/div[2]/div/div/dl[2]/dd/span').text)
            info.append((driver.find_element(By.XPATH,'//*[@id="app"]/div/div[3]/div[4]/div/div[2]/div/div/dl[3]/dd').text).split('\n'))
            info.append((driver.find_element(By.XPATH,'//*[@id="app"]/div/div[3]/div[4]/div/div[2]/div/ul').text).split('\n'))
            driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div[4]/div/div[3]/button').click()
            time.sleep(0.4)
            get_column += info
except:
    pass
        
```
